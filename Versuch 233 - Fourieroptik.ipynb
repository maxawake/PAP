```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

def beugung_spalt(x):
    return np.sinc(x)**2

def beugung_doppelspalt(x):
    return np.sinc(x)**2*np.cos(np.pi*v*x)**2

#Anzahl a der Nebenmaxima die dargestellt werden sollen:
a=2
#Verhältnis Spaltabstand zu Spaltbreite
v=2.325

x=np.linspace(-(a+1),a+1,2000)

plt.plot(x,beugung_spalt(x),label="Spalt", linestyle="dashed", color="grey")
plt.plot(x,beugung_doppelspalt(x),label="Doppelspalt", color="#9516E0")
plt.xlabel("$\pi x$")
plt.ylabel("Intensität [b.E.]")
plt.title("Abb.A.3: Theoretisches Beugungsbild des Doppelspalts")
plt.ylim((0,1.1))
plt.legend()
plt.savefig("233_3.pdf", format="pdf")
```

```python
#Integrand
def spalt(k):
    return d/np.pi*np.sin(k*d/2)/(k*d/2)*np.cos(y*k)

d=1 #Spaltbreite
n=3 #Integrationsbereich

#y-Werte
Y=np.linspace(-1,1,200)*d

#Berechnung des resultierenden Objektbilds
from scipy.integrate import quad
f_mod=[]
for i in range(len(Y)):
    y=Y[i]
    result, error = quad(spalt,0,2*np.pi*n/d)
    f_mod.append(result**2)
    
#Kurve auf 1 normieren:
f_mod=f_mod/np.max(f_mod)

plt.plot(Y,f_mod, color="#9516E0")
plt.xlabel("y [d]")
plt.ylabel("Intensität [b.E.]")
#plt.title("Abb.A.4: Theoretisches Objektbild des Modifizierten Spalts")
plt.ylim(0,1.1)
plt.savefig("233_4_3.pdf", format="pdf")
```
