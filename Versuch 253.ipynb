```python
%matplotlib inline 
import matplotlib.pyplot as plt
import numpy as np
from IPython.display import display, Math, Latex

d = np.array([0, 0.3, 0.6, 0.9, 1.2, 1.5, 1.8, 2.1, 2.4, 2.7, 3.0, 3.3, 3.6, 4.6])
n = np.array([1615, 1000, 660, 449, 313 ,192, 123, 257/4, 133/4, 96/4, 80/4, 59/4, 57/4, 128/10])
ns = (n-(128/10))/4

from scipy.optimize import curve_fit
def exp(x,a,l):
    #return a**2 * np.exp( -l * x + np.sqrt(1615/4) ) 
    return a*x+l

popt, pcov=curve_fit(exp, d[8:12], ns[8:12]) #p0=[2000, 0.02])
x = np.linspace(0,5,100)
plt.plot(x, exp(x,*popt), label="Extrapolation", color="orange", linewidth=1)

popt_2, pcov_2=curve_fit(exp, d[8:12], ns[8:12]) #p0=[2000, 0.02])
plt.plot(x, exp(x-0.5,*popt_2)-1, label="Fehlergerade", linestyle="--", linewidth=1, color="gray")

plt.plot(d, ns, ".", color="black", label="Messdaten")
plt.errorbar(d, ns, np.sqrt(ns), fmt=".", markersize=1, capsize=2, label="Fehler", color="black")
plt.title("Abb.3: Absorption von β-Strahlung in Aluminium")
plt.ylabel("korrigierte Zählrate (n-n_0_β) (log)")
plt.xlabel("Aluminiumdicke [mm]")
plt.yscale("log")
plt.legend()
plt.savefig("C:/Users/Max/int4.pdf", format="pdf")
```

```python
#Hier werden Chi^2 Summen und reduzierte Chi^2 Summen ausgerechnet 
chi2_g=np.sum((exp(d[7:13], *popt)-ns[7:13])**2/np.sqrt(ns[7:13]))
dof_g=len(d[7:13])-2 #dof: degrees of freedom, Freiheitsgrade
chi2_red_g=chi2_g/dof_g
print("chi2_g=", chi2_g)
print("chi2_red_g",chi2_red_g)
from scipy.stats import chi2
prob_g=round(1-chi2.cdf(chi2_g,dof_g),3)*100
print("Wahrscheinlichkeit=", prob_g,"%")
```
