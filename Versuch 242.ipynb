```python


u_l = np.array([250,190,130,60,-60,-130,-190,-250])
u_a1 = np.array([-4.08,-3.10,-2.11,-0.98,0.98,2.11,3.08,4.06])
u_a2 = np.array([-13.1,-13.1,-11.9,-5.45,5.55,11.9,14.3,14.3])
x = np.linspace(-300,300,100)
du_l = 0
du_a1 = 0.03
du_a2 = 0.1

from scipy.optimize import curve_fit 

def lin(x,a,b):
    return a*x+b

popt, pcov=curve_fit(lin,u_l, u_a1)
plt.plot(x, lin(x,*popt), label='Extrapolation m={}'.format(round(popt[0]*1000,3)), color="#B14AE9",linestyle="-")

plt.errorbar(u_l,u_a1,du_l,du_a1, marker=".", linestyle="none", label="Messdaten".format(round(popt[0],3)), color="black")

plt.title("Abb.8: Verhältnis der Ein- und Ausgangsspannung \nbei Gleichspannung mit R=48,7 k$\Omega$")
plt.xlabel("Eingangsspannung $U_E [mV]$")
plt.ylabel("Ausgangsspannug $U_A [V]$".format(round(popt[0],3)))
plt.legend()
print(popt[0]*1000)
plt.savefig("C:/Users/Max/dropbox/Uni/python/v242_5.pdf", format="pdf")
```
