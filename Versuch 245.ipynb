```python
%matplotlib inline 
import matplotlib.pyplot as plt
import numpy as np
from scipy.optimize import curve_fit

us = np.array([0.95,1.95,3.05,4.08,5.15])
w = np.array([3.0,5.9,9.0,12.0,15.0])
x = np.linspace(0,15.5,100)
u = us/np.sqrt(2)
def f(x,a,b):
    return b+a*x
popt, pcov=curve_fit(f, w, u)
plt.plot(x, f(x,*popt), label="Extrapolation", color="limegreen",linestyle="-")
plt.plot(w, u, ".", color="black", label="Messdaten")
plt.errorbar(w, u, 0.1, 0.1, fmt=".", markersize=1, capsize=2, label="Fehler", color="black")
plt.title("Abb.5a: Induktionsspannung in Abbhängigkeit der Frequenz")
plt.ylabel("Spannung $U_eff$ [V]")
plt.xlabel("Frequenz f [1/s]")
print(popt[0],np.sqrt(popt[0]))
plt.legend()

plt.savefig("C:/Users/Max/245_1.pdf", format="pdf")
```
