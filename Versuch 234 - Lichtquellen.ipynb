```python
%matplotlib inline
import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import numpy as np

def comma_to_float(valstr):
    return float(valstr.decode("utf-8").replace(',','.'))

lamb_og, inten_og=np.loadtxt("Himmelslicht_ohne_fenster.txt", skiprows=17, converters={0:comma_to_float,1:comma_to_float}, comments=">", unpack=True)
lamb_mg, inten_mg=np.loadtxt("Himmelslicht_mit_fenster.txt", skiprows=17, converters={0:comma_to_float,1:comma_to_float}, comments=">", unpack=True)

plt.plot(lamb_og,inten_og, label="Ohne Glas",color="black", linewidth=1)
plt.plot(lamb_mg,inten_mg, label="Mit Glas",color="orange", linewidth=1)
plt.legend()
plt.title("Abb.A.1: Sonnenspektrum mit und ohne Glas")
plt.xlabel("Wellenlänge $\lambda$ [nm]")
plt.ylabel("Intensität [b.E.]")
plt.grid(linestyle="dotted")
plt.xlim(300,800)
plt.savefig("234_1.pdf", format="pdf")
```

```python
A = 1 - ((inten_mg)/inten_og)
plt.plot(lamb_og,A,color="black", linewidth=1)
plt.xlim(300,800)
plt.ylim(-0.2,1.3)
plt.title("Abb.A.2: Absorption von Glas")
plt.xlabel("Wellenlänge $\lambda$ [nm]")
plt.ylabel("Absorption [b.E.]")
plt.savefig("234_2.pdf", format="pdf")
```

```python
from scipy.optimize import curve_fit
from scipy.stats import chi2

wellenl = np.array([818.7,567.6,497.2,465.9,448.9,438.3,432.0,425.8,419.1])
fehler = np.array([1.5,1.0,1.2,1.2,1.2,0.8,1.0,1.0,1.1])
quantz = np.array([3,4,5,6,7,8,9,11,12])
d = np.linspace(2.8,13,100)

def fit(m,E_Ry,E_3p,D_d):
    return 1.2398E3/(E_Ry/(m-D_d)**2-E_3p)
para = [-13.6,-3,-0.02]

popt,pcov = curve_fit(fit,quantz,wellenl, sigma=fehler, p0=para)
plt.plot(d,fit(d,*popt), label="Fit",color="orange")

chi2_ = np.sum((fit(quantz,*popt)-wellenl)**2/fehler**2)
dof = len(quantz)-3
chi2_red = chi2_/dof
print("chi2=", chi2_)
print("chi2_red=", chi2_red)


prob = round(1-chi2.cdf(chi2_,dof), 2)*100
print("Wahrscheinlichkeit:", prob, "%")

plt.plot(quantz,wellenl,color="black",marker="+", label="Messwerte", linestyle="none")

plt.title("Abb.A.7: 1. Nebenserie des Na-Atoms")
plt.xlabel("Quantenzahl m")
plt.ylabel("Wellenlänge [nm]")
plt.legend()

print("E_Ry=", popt[0],"stder=", np.sqrt(pcov[0][0]))
print("E_3p=", popt[1],"stder=", np.sqrt(pcov[1][1]))
print("D_d=", popt[2],"stder=", np.sqrt(pcov[2][2]))

plt.savefig("234_7.pdf", format="pdf")
```
